```python
import pandas as pd
import matplotlib.pyplot as plt  # Librería para crear gráficos
import numpy as np  # Librería para generar datos numéricos, aquí la usamos para generar datos aleatorios
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Dannap7337/Pato/main/Base_limpiaclase.csv')
df=df.drop(columns=['Unnamed: 0'])
df
```

```python
df.columns
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(df['Balance'], fill=True, color="red")
plt.title("Grafica de balance")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(df['Salario'], fill=True, color="red")
plt.title("Grafica de salario")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(df['CreditScore'], fill=True, color="blue")
plt.title("Grafica de creditscore")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.kdeplot(df['Edad'], fill=True, color="green")
plt.title("Grafica de edad")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# KDE para Balance
sns.kdeplot(df['Balance'], fill=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribución de Balance')

# KDE para Salario
sns.kdeplot(df['Salario'], fill=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribución de Salario')

# KDE para CreditScore
sns.kdeplot(df['CreditScore'], fill=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribución de CreditScore')

# KDE para Edad
sns.kdeplot(df['Edad'], fill=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribución de Edad')

# Ajustar el layout para evitar superposición de títulos y etiquetas
plt.tight_layout()

# Mostrar gráfico
plt.show()
```

```python
gender_counts = df['Genero'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90,
         colors=['skyblue', 'lightgreen'])
plt.title('Distribución por Género')
plt.show()
```

```python
pais_counts = df['Pais'].value_counts()

# Gráfico de pastel
plt.figure(figsize=(6,6))
plt.pie(pais_counts, labels=pais_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'skyblue', 'lightgreen', 'yellow'])
plt.title('Distribución por País')
plt.show()
```

```python
# Gráfico de violín para CreditScore por Tipo de Tarjeta
sns.violinplot(x='CreditScore', data=df)

# Mostrar gráfico
plt.title('Distribución de CreditScore')
plt.show()
```

```python
colu=[ 'CreditScore', 'Edad',
       'Estadia', 'Balance', '#Prod', 'TarjetaCredito', 'Activo', 'Salario',
       'Salida', 'Quejas', 'Nivel_satisf', 'Puntos']
```

```python
for i in colu:
    # Gráfico de violín para CreditScore por Tipo de Tarjeta
    sns.violinplot(y=i, data=df)

    # Mostrar gráfico
    plt.title(f'Distribución de {i}')
    plt.show()
```
